```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

1. Understand the Data Structure

```python
df.info()
```

2. Summarized Statastics

```python
df.describe()
```

3. Data Visualization

3.1. Histograms

```python
df.hist(bins=20, figsize=(10, 8))
plt.show()
```

3.2. Pair Plot

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df)
plt.show()
```

```python
sns.boxplot(df, x='species', y='sepal_length')
plt.show()

sns.boxplot(df, x='species', y='sepal_width')
plt.show()

sns.boxplot(df, x='species', y='petal_length')
plt.show()

sns.boxplot(df, x='species', y='petal_width')
plt.show()
```

```python
sns.violinplot(df, x='species', y='sepal_length')
plt.show()

sns.violinplot(df, x='species', y='sepal_width')
plt.show()

sns.violinplot(df, x='species', y='petal_length')
plt.show()

sns.violinplot(df, x='species', y='petal_width')
plt.show()
```

```python
clm = df.select_dtypes(include=[float,int])
corr_matrix = clm.corr()
#heat map
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
